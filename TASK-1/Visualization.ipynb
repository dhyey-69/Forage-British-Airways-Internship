Imported libraries we are going use in data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Read our data file / csv file from drive and store it inside data variable

```python
data = pd.read_csv('/content/drive/MyDrive/Internship/Sem-5/TASK-1/cleaned_data.csv')
data.head()
```

According to data we have reviews from 70 different countries

```python
print(f"{len(data.country.unique())} unique countries")
```

Average Overall Rating is 4.6 Stars out of 10 stars

```python
data['stars'].mean()
```

There are only 65.8% reviews which is having verified trip and rest 34.2% are not verified trip.

```python
d = data.verified.value_counts()
l = ['Verified', 'Unverified']
plt.pie(d, labels=l,autopct='%1.1f%%',startangle=90)
plt.title('Percentage of Verified and Not Verified Trips')
plt.show()
```

Out of all reviews given we have total 69.3% reviews positive,30.2% reviews are negative also,so we are looking forward to reduce this 30.2% numbers, We have 0.5% reviews which are both positive and neagtive also

```python
d = data.sentiment.head(1000).value_counts()
l = ['Positive', 'Negative', 'Neutral']
plt.pie(d, labels=l,autopct='%1.1f%%',startangle=90)
plt.title('Types of Reviews')
plt.show()
```

Top 10 countries from where we get reviews.

```python
d1 = pd.DataFrame(data['country'])
count = data.country.value_counts()
top = count.head(10)
print(top)
```

```python
plt.figure(figsize=(10, 6))
top.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Number of Reviews')
plt.xlabel('Country')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#Star Values Distribution

```python
data.stars.value_counts()
```

```python
plt.figure(figsize=(10, 10))
plt.hist(data['stars'], bins=range(1, 12), align='left', edgecolor='black', color='skyblue',width=0.8)
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Number of Trips')
plt.xticks(range(1, 11))
plt.show()
```

```python
star_counts = data['stars'].value_counts().sort_index()

plt.figure(figsize=(8, 8))
plt.pie(star_counts, labels=star_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Star Ratings')
plt.show()
```
